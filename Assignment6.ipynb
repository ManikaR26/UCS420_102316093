```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='$y = Mx^2$', color='b', linestyle='-')
plt.plot(x, y2, label='$y = Msin(x)$', color='r', linestyle='--')

plt.legend()
plt.grid(True)
plt.title(f'Graph of Mathematical Function for M={M}')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

subjects = ['Physics', 'Chemistry', 'Biology', 'Mathematics', 'Computer Science']
scores = [87, 92, 88, 91, 94]
df = pd.DataFrame({'Subject': subjects, 'Score': scores})

sns.set_style("whitegrid")

plt.figure(figsize=(10, 5))
barplot = sns.barplot(x='Subject', y='Score', data=df, palette='viridis')

barplot.bar_label(barplot.containers[0], fmt='%d', fontsize=12, color='black', label_type='edge')

plt.title("Scores in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 102316093
np.random.seed(roll_number)
data = np.random.randn(50)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(np.cumsum(data), marker='o', linestyle='-', color='b')
axes[0, 0].set_title("Cumulative Sum Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True, linestyle='--', alpha=0.6)

axes[0, 1].scatter(range(50), data, color='r', alpha=0.7)
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value")
axes[0, 1].grid(True, linestyle='--', alpha=0.6)

fig.delaxes(axes[1, 0])
fig.delaxes(axes[1, 1])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv")


plt.figure(figsize=(9, 5))
plt.plot(df['month_number'], df['total_profit'], marker='o', linestyle='-', color='b', label='Total Profit')
plt.title("Total Profit Over Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()


plt.figure(figsize=(9, 5))
for column in df.columns[1:-2]:
    plt.plot(df['month_number'], df[column], marker='s', linestyle='--', label=column)
plt.title("Product Sales Over Months")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Bar chart for all attributes
df[df.columns.difference(['month_number'])].plot(kind='bar', figsize=(12, 6), width=0.8, edgecolor='black')
plt.title("All Attributes")
plt.xlabel("Month Number")
plt.ylabel("Values")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


```
