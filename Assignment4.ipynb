```python
def most_frequent(array):
    unique, counts = np.unique(array, return_counts=True)  
    most_frequent_value = unique[np.argmax(counts)] 
    indices = np.where(array == most_frequent_value)[0]  
    return most_frequent_value, indices


x = np.array([1, 2, 3, 4, 5, 1, 2, 1, 1])
most_frequent_x, indices_x = most_frequent(x)
print("Most frequent value in x:", most_frequent_x)
print("Indices:", indices_x)


y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
most_frequent_y, indices_y = most_frequent(y)
print("Most frequent value in y:", most_frequent_y)
print("Indices:", indices_y)
```

```python
ucs420_array = np.array([[10, 20, 30, 40], 
                         [50, 60, 70, 80], 
                         [90, 15, 20, 35]])
mean_value = np.mean(ucs420_array)
median_value = np.median(ucs420_array)
max_value = np.max(ucs420_array)
min_value = np.min(ucs420_array)
unique_elements = np.unique(ucs420_array)
reshaped_ucs420_Manika = ucs420_array.reshape(4, 3)
resized_ucs420_Manika = np.resize(ucs420_array, (2, 3))
print("Original 2D Array (3x4):\n", ucs420_array)
print("\nMean:", mean_value)
print("Median:", median_value)
print("Max:", max_value)
print("Min:", min_value)
print("Unique Elements:", unique_elements)
print("\nReshaped Array (4x3):\n", reshaped_ucs420_Manika)
print("\nResized Array (2x3):\n", resized_ucs420_Manika)
```
