```python
import numpy as np

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print("The matrix is :", gfg)
```

```python
sumall = np.sum(gfg)

rowsum = np.sum(gfg, axis = 1)

columnsum = np.sum(gfg, axis = 0)

print("Sum of all elements:", sumall)
print("Row-wise sum:\n", rowsum)
print("Column-wise sum:", columnsum)
```

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])
sortedarray = np.sort(array)
print("Sorted Array:", sortedarray)

sortedindices = np.argsort(array)
print("Indices of Sorted Array:", sortedindices)

foursmallest = np.sort(array)[:4]
print("4 Smallest Elements:", foursmallest)

fivelargest = np.sort(array[-5:])
print("5 Largest Elements:", fivelargest)
```

```python
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
integer_elements = array[array == array.astype(int)]
print("Integer only:",integer_elements)
float_elements = array[array != array.astype(int)]
print("Float only:",float_elements)
```

```python
X = ord('A')+ord('G')
sales = np.array([X, X+50, X+100, X+150, X+200])
print(sales)
```

```python
taxrate = ((X % 5) + 5)/ 100

taxedsales = sales * (1 - taxrate)
print("Sales after tax deduction:", taxedsales)
```

```python
discountedsales = np.where(sales < X + 100, sales * 0.95, sales * 0.90)
print("Discounted sales:", discountedsales)
```

```python
salesmatrix = np.tile(sales, (3, 1))
weeklyincrease = np.array([1.00, 1.02, 1.04]).reshape(3,1)
adjustedsales = salesmatrix * weeklyincrease

print("Weekly Sales Matrix:", adjustedsales)
```

```python
X = np.linspace(-10, 10,100)
print(f"X array {X}")
Y1 = np.power(X, 2)
Y2 = np.sin(X)
Y3 = np.exp(X)
Y4 = np.log(np.abs(X) + 1)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(X, Y1, label=r'$y = x^2$', color='b')
plt.plot(X, Y2, label=r'$y = \sin(x)$', color='g')
plt.plot(X, Y3, label=r'$y = e^x$', color='r')
plt.plot(X, Y4, label=r'$y = \log(|x| + 1)$', color='m')
plt.title("Plots of Different Mathematical Functions")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```
